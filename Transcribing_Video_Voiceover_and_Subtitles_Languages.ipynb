```python
import whisper

video_file = "PIKSEL.mp4"

model = whisper.load_model("base")
result = model.transcribe(video_file, task="translate")

# Save full English transcript
with open("PIKSEL_EN.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

print("Transcription & translation completed.")
```

```python
def format_timestamp(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = seconds % 60
    return f"{hours:02}:{minutes:02}:{seconds:06.3f}".replace('.', ',')

with open("PIKSEL_EN.srt", "w", encoding="utf-8") as f:
    for i, seg in enumerate(result["segments"], start=1):
        f.write(f"{i}\n")
        f.write(f"{format_timestamp(seg['start'])} --> {format_timestamp(seg['end'])}\n")
        f.write(seg["text"].strip() + "\n\n")

print("Subtitle file created: PIKSEL_EN.srt")
```

```python
from gtts import gTTS

tts = gTTS(result["text"], lang="en")
tts.save("PIKSEL_EN_voice.mp3")

print("English voice-over audio created.")
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip

video = VideoFileClip(video_file)
audio = AudioFileClip("PIKSEL_EN_voice.mp3")

final_video = video.set_audio(audio)
final_video.write_videofile(
    "PIKSEL_EN_DUBBED.mp4",
    codec="libx264",
    audio_codec="aac"
)

print("Dubbed video created.")
```
